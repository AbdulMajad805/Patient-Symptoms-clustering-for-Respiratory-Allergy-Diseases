```python
# Stats and Prevalence:
# To know how common each symptom is. For binary features, the mean × 100 = prevalence %.
```

```python
symptom_cols = [col for col in data.columns]

# Prevalence of each symptom
symptom_summary = data[symptom_cols].mean().sort_values(ascending=False) * 100
symptom_summary = symptom_summary.round(2).reset_index()
symptom_summary.columns = ["Symptom", "Prevalence (%)"]

print(symptom_summary)
```

```python
# Tells me: “COUGH occurs in 52% of patients, TIREDNESS in 48%, etc.”
# “High prevalence symptoms = COUGH, RUNNY_NOSE, … Low prevalence = ITCHY_ etc.”
```

```python
# Value Counts for Each Symptom:
```

```python
for col in symptom_cols:
    print("\n",data[col].value_counts(normalize=True) * 100)
```

```python
# Symptom Co-occurrence:
```

```python
from itertools import combinations
from collections import Counter

co_occurrence = Counter()

for comb in combinations(symptom_cols, 2):
    pair_count = ((data[comb[0]] == 1) & (data[comb[1]] == 1)).sum()
    co_occurrence[comb] = pair_count

# Convert to DataFrame
co_occurrence_df = pd.DataFrame(co_occurrence.items(), columns=["Symptom_Pair", "Count"])
co_occurrence_df = co_occurrence_df.sort_values("Count", ascending=False).head(15)

print(co_occurrence_df)
```

```python
# To see which symptoms often appear together.
# We’ll count co-occurrence for every pair.
```

---
**Symptom_Pair** → two symptoms that appear together in patients.

**Count** → how many patients had both of these symptoms.

- For example:
**(COUGH, TIREDNESS) = 11,175 patients had both cough and tiredness.**
---

```python
# Checking Coorelation:
```

```python
import seaborn as sns

corr = data[symptom_cols].corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Symptom Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Checking Association rule:
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Convert to True/False
df_bool = data[symptom_cols].astype(bool)

# Find frequent itemsets
frequent_items = apriori(df_bool, min_support=0.1, use_colnames=True)

# Generate rules
rules = association_rules(frequent_items, metric="lift", min_threshold=1.0)
rules = rules.sort_values("confidence", ascending=False).head(10)

print(rules[["antecedents", "consequents", "support", "confidence", "lift"]])
```

---
**Antecedents** = “IF” part (symptom(s) on the left side of the rule)

**Consequents** = “THEN” part (symptom(s) predicted on the right side)

**Support** = % of patients who have both antecedent and consequent

- Example: (LOSS_OF_TASTE) → (COUGH) has support 0.2373 = ~23.7% of all patients have both loss of taste & cough.

**Confidence** = probability of consequent given antecedent

- Example: For patients with LOSS_OF_TASTE, 53.4% also had COUGH.

**Lift** = how much more often the rule happens compared to random chance

- **Lift > 1** = positive association (they occur together more than random).
- **Lift ~1** = no strong association.
- **Lift < 1** = negative association (rarely occur together).
---

```python
# Lets do PCA:
    # it nedds scaled data , so...
```

```python
X= data
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Now PCA for noise reduction & visualization:
    # Use PCA to reduce to 2D or 3D for quick visualization.
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Fit PCA with all components
pca = PCA(n_components=len(X.columns))
X_pca_full = pca.fit_transform(X_scaled)
```

```python
pca.fit_transform(X_scaled)
```

```python
# Variance explained
explained_var = np.cumsum(pca.explained_variance_ratio_)
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(explained_var, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.show()
```

```python
# Reduce to 2D for plotting
pca2 = PCA(n_components=2)
X_pca = pca2.fit_transform(X_scaled)
pca2.fit_transform(X_scaled)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.5)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA (2D) Projection of Symptom Data")
plt.show()
```

---
- Each dot = one patient (with their symptoms encoded).
- **PC1 (x-axis)** = the direction of maximum variance in your dataset (the strongest symptom pattern).
- **PC2 (y-axis)** = the second most important direction of variance, orthogonal to PC1.

So instead of plotting **30+ symptoms**, PCA compresses them into **2 “summary features” (PC1, PC2)** that capture most of the structure.

---

```python
#  Now lets do t-SNE:
    # After PCA, run t-SNE on the scaled data (sometimes also on top 30 PCA components for speed).
    # 👉 This gives a 2D “map” of patients, where clusters are usually clearer.
```

```python
from sklearn.manifold import TSNE
sample= X_scaled[:5000]
# Run t-SNE on PCA-reduced data (say 30 comps) or directly on X_scaled
pca20 = PCA(n_components=20).fit_transform(sample)

tsne = TSNE(n_components=2, perplexity=30, max_iter=500,random_state=42)
X_tsne = tsne.fit_transform(pca20)
tsne.fit_transform(pca20)
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], alpha=0.5)
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.title("t-SNE Projection of Symptom Data")
plt.show()
```

---
- Each dot = one patient (row) in your dataset.
- The **x-axis (t-SNE 1)** and **y-axis (t-SNE 2)** are **artificial dimensions** created by t-SNE — they don’t correspond to a real symptom.
- Dots that are **close together** = patients with **similar symptom patterns.**
- Dots that are **far apart** = patients with very **different symptom combinations.**

If your dataset has **natural clusters** (like flu-like, allergy-like, COVID-like, GI-like symptom groups), t-SNE will often make them **visibly separate blobs** in the scatter plot.

- You might see **several dense “clouds” of points**. Each cloud = a symptom cluster.
- If clusters overlap a lot → your symptoms are not strongly separable.
---

```python
# If I color the points by KMeans cluster label, I'll see which group belongs to which cluster. Example:
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)  # try 3–6 clusters and compare (just for comparison, not actual model..)
kmeans_labels = kmeans.fit_predict(pca20)

# =====================
# Plot with cluster colors
# =====================
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_tsne[:,0], X_tsne[:,1], 
                      c=kmeans_labels, cmap="tab10", alpha=0.6)

plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.title("t-SNE Projection with KMeans Clusters")

# add color legend
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```
